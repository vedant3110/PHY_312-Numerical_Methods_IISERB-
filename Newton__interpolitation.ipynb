```python
import numpy as np
import scipy as sc
import math as ma
import matplotlib.pyplot as plt
```

```python
x=np.array([1,4,5,6])
y=np.array([0.0,1.38629401,1.60943794,1.79175901])
```

```python
n=4
m=int(input("Enter degree of polinomial"))
```

```python
if m>=n:
  print("The polynomial cannot be interpolitated")
else:
  f=np.zeros([n,n])
  f[:,0]=y
  for i in range(1,n):
    for j in range(0,n-i):
      f[j,i]=(f[j+1,i-1]-f[j,i-1])/(x[i+j]-x[j])
  x_pre=np.linspace(1,6,20)
  y_pre=np.zeros(20)

  for i in range(0,20):
    coeff=1
    temp=f[0,0]
    for k in range(0,m):
      coeff*=(x_pre[i]-x[k])
      temp+=coeff*f[0,k+1]
    y_pre[i]=temp
  plt.plot(x_pre,y_pre,"r",label="Interpolitated Data")
  plt.scatter(x,y,color="g",label="Data")
  plt.legend()
  plt.grid()
  plt.show()
```
