```python
import numpy as np
import matplotlib.pyplot as plt
import math as mth
import scipy as sc
```

```python
freq = 1.0
Fs = 100
T = 1/freq
N = int(Fs * T)
time = np.linspace(-T/2, T/2, N, endpoint=False)
```

```python
m = int(input('Enter order of the series (m) ='))
```

```python
def f(t,m):
  temp=np.zeros(len(t),dtype="float64")
  for i in range(1,m+1):
    temp+=((-1)**(i+1)*4*np.cos(2*np.pi*(2*i-1)*freq*t))/((2*i-1)*np.pi)
  return temp
func=f(time,m)
plt.plot(t,func)
```

```python
fft_result = np.fft.fft(func)
dt = time[1] - time[0]
frequencies = np.fft.fftfreq(N, dt)
amplitude = np.abs(fft_result) / len(time)
plt.stem(frequencies[0:N//2], amplitude[0:N//2], 'r', markerfmt="bo")
```
