#Newton-Raphson Method

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
import math as ma
```

Procedure to do it

image.png

Step 1:Choose a intial point   $x_i$

Step 2:Compute:
$$
  x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$
Step 3:Error Analysis:
$$
  E_{t, i+1} \approx \frac{f''(x_r)}{2 f'(x_r)} (E_{t, i})^2
$$

###Termination Criteria and Error Estimates

an approximate percent relative error  ea :
$$
  e_a=\left |\frac{x_r^{new}-x_r^{old}}{x_r^{new}}\right |\cdot100 \%
$$

```python
def f(x):
  return ma.e**x-10##put function of your choice
```
