###Termination Criteria and Error Estimates

*  an approximate percent relative error $e_a$:



$$
  e_a=\left |\frac{x_r^{new}-x_r^{old}}{x_r^{new}}\right |\cdot100 \%
$$
*  the root must lie within $±\Delta x/2$  [$\Delta x=x_r^{new}-x_r^{old}$]
*  A method such as false position is often superior
to bisection, there are invariably cases that violate this general conclusion.



```python
def f(x):
  return ma.e**x-10##put function of your choice
```

```python
t=np.linspace(0,4)
f_x=[]
for i in t:
  f_x.append(f(i))
f_x=np.array(f_x)
plt.plot(t,f_x)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(y=0, color="black", linestyle=":")
```

predicted value of $x_l=1.5$,and $x_u=3.5$

```python
x_l=1.5;
x_u=3.5;
e_a=100;
x_r=x_u-(f(x_u)*(x_u-x_l)/(f(x_u)-f(x_l)))
itr=0;
itr_max=100;
while(e_a>0.0001 and itr<itr_max):
  if f(x_l)*f(x_r)<0:
    x_u=x_r
  elif f(x_u)*f(x_r)<0:
    x_l=x_r
  elif f(x_l)*f(x_r)==0:
    break
  x_r_new=x_u-(f(x_u)*(x_u-x_l)/(f(x_u)-f(x_l)))
  e_a=abs((x_r_new-x_r)/x_r_new)*100
  error=abs(x_r_new-x_r)/2
  print(f"lower limit: {x_l:<15.6f} high limit: {x_u:<10.1f} mean: {x_r:<15.6f} new mean: {x_r_new:<15.6f} error: {e_a:<15.6f}")
  x_r=x_r_new
  itr+=1

print(f"\n\nFinal answer:{x_r:.6f} ± {error:.6f}")
```
