```python
import matplotlib.pyplot as plt
import math as ma
import numpy as np
import scipy as sc
```

```python
x = np.linspace(2,15,15)
y = np.array([10,16.3,23,27.5,31,35.6,39,41.5,42.9,45,46,45.5,46,49,50])
```

```python
plt.scatter(x,y)
```

```python
n=15
m=4
```

```python
zt=np.zeros([n,m+1])
zt[0,1]=2
```

```python
for i in range(0,n):
  for j in range(0,m+1):
    zt[i,j]=x[i]**j
```

```python
Z=np.matmul(np.transpose(zt),zt)
Y=np.matmul(np.transpose(zt),y.T)
```

```python
A=np.linalg.solve(Z,Y).ravel()
```

```python
x_pre=np.linspace(2,15,50)
y_pre=np.zeros(50,dtype=float)
for i in range(0,50):
  for j in range(0,m+1):
    y_pre[i]+=A[j]*x_pre[i]**j
```

```python
plt.scatter(x,y,color='green',label="Experimental Data")
plt.plot(x_pre,y_pre,'r',label="Curve-fit Data")
plt.xlabel="x"
plt.ylabel="y"
plt.legend()
plt.grid()
```
